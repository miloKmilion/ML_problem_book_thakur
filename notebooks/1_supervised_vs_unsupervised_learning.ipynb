### 1. Supervised vs unsupervised learning

Supervised data: Always has one or multiple targets associated with it.
Unsupervised data: Does not have any target variable.

Notes:
* A supervised problem is considerable easier to tackle than an unsupervised one.
* A problem in where is required to predict a value is known a supervised problem.
* Classify images is also supervised.

#### Differences by target type:
* If the target is categorical, the problem becomes a classification problem
* If the target is a real number, then the problem is defined as a regression problem.

### Unsupervised learning:
This type of problems the datasets do not have a target associated with them and in general are more challenging

Example: Identifying credit card transactions as valid or fraud, As there is no information about a transaction being
genuine or not, hence, the problem becomes unsupervised.

To tackle these problems it is important to think how many clusters can the data be divided into
* Clustering is one of the approaches used for problems like this. Alongside other approaches valid for unsupervised problems.
* After clustering is applied, and the number of clusters is known, it is necessary to be able to distinguish between the assumed targets.
* To make sense of unsupervised problems, decomposition techniques such as PCA, t-SNE can be used

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

from sklearn import datasets
from sklearn import manifold

%matplotlib inline
```

```python
# Importing the datasets
data = datasets.fetch_openml('mnist_784', version=1, return_X_y=True)

pixel_values, targets = data
targets = targets.astype(int)
```

The data loaded form sklearn, is an array of pixel values and another arrays of targets, the targets were originally strings and needed to be converted to integer

pixel_values is a 2-dimensional array of shape 70000 x 784. this mean 70000 different images, each of them of size 28 x 28 pixels. After flattening will give a total of 784 data points

```python
# Reshaping the image to a 28 x 28 matrix and then plotting it.
pixel_values = np.array(pixel_values)
single_image = pixel_values[1, : ].reshape(28,28)
plt.imshow(single_image, cmap="gray")
```

```python
# Reducing the dimensionality of the data from 784 to 2.
tsne = manifold.TSNE(n_components=2, random_state=42)
transformed_data = tsne.fit_transform(pixel_values[:3000, :])
```

The step above, creates a t-SNE transformation of the data. By using 2 components it is possible to visualize in a 2-dimensional
setting.
The transformed_data is an array of shape 3000 x 2. In the next step it will be converted to a pandas dataframe.

```python
# Creating a dataframe from a numpy array.
tsne_df = pd.DataFrame(np.column_stack((transformed_data, targets[:3000])), columns=["x", 'y', 'targets'])

tsne_df.loc[:, "targets"] = tsne_df.targets.astype(int)

tsne_df.head(10)
```

The pandas dataframe was created from a numpy array. X and Y are the components from t-SNE decomposition and targets is the actual
number.

```python
grid = sns.FacetGrid(tsne_df, hue="targets", height=8)
grid.map(plt.scatter, "x", "y").add_legend()
```

This is one example of how to visualize unsupervised datasets.
